```python
def geocode(addr_string):
    res = gmaps.geocode(addr_string)
    latlng = res[0]['geometry']['location']
    lat = latlng["lat"]
    lng = latlng["lng"]
    addr = res[0]['formatted_address']
    return {'lat':lat,"lng":lng, "addr":addr}
```

```python
def df_geocode(csv_path):
    addr = pandas.read_csv(csv_path)
    addr = addr[['full_addr', 'apn_sort']]
    addr10 = addr[0:10]
    addr10["geo"] = [geocode(i) for i in addr10["full_addr"]]
    addr10['lat'] = [i['lat'] for i in addr10["geo"]]
    addr10['lng'] = [i['lng'] for i in addr10["geo"]]
    addr10['addr'] = [i['addr'] for i in addr10["geo"]]
    return addr10
```
