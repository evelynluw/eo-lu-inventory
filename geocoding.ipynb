```python
def centroidCoords(shapeString):
    x=ast.literal_eval(shapeString)
    #flipped order to achieve standard format of (lat,lng) 
    flippedCoords = list(shp.shape(x).centroid.coords)[0]
    return (flippedCoords[1],flippedCoords[0])
```

```python
#input csv file path and name of column which contains the polygon data. 
def df_revgeocode(csv_path, col):
    poly = pandas.read_csv(csv_path)
    #only testing on first 10 rows to minimize api usage
    poly10 = poly[:10]
    poly10 = poly10[[col]]
    #remove rows without polygon data
    cleaned = poly10[poly10.isna()[col]==False]
    #compute centroid of each polygon
    centroids = [centroidCoords(i) for i in cleaned[col]]
    cleaned["centroids"] = centroids
    #reverse geocode to find matching addresses for centroid coordinates
    #some have multiple matching addresses for their coordinate => ouput as list
    cleaned["full_addrs"]=[gmaps.reverse_geocode(i,result_type="street_address") for i in cleaned["centroids"]]
    cleaned["formatted_addrs"] = [[j['formatted_address'] for j in i] for i in cleaned["full_addrs"]]
    return cleaned
```
